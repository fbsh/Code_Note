```r
finals.table = read.table("Table3.10.txt", header=TRUE, sep='\t')
plot(finals.table, pch=23, bg='orange', cex=2)
```

```r
finals.lm = lm(F ~ P1 + P2, data=finals.table)
finals.lm2 = lm(F ~ P1, data=finals.table)
finals.lm3 = lm(F ~ P2, data=finals.table)
summary(finals.lm)
```

```r
par(mfrow=c(2,2))
plot(finals.lm, pch=23 ,bg='orange',cex=2)
```

## Model 1

```r
plot(fitted(finals.lm), resid(finals.lm), pch=23, bg='red', cex=2)
abline(h=0, lty=2)
```

## Model 2

```r
plot(fitted(finals.lm2), resid(finals.lm), pch=23, bg='red', cex=2)
abline(h=0, lty=2)
```

## Model 3

```r
plot(fitted(finals.lm3), resid(finals.lm), pch=23, bg='red', cex=2)
abline(h=0, lty=2)
```

# Outlying  X  values
leverage $_{i}=H_{i i}=\left(X\left(X^{T} X\right)^{-1} X^{T}\right)_{i i}$
## Model 1

```r
plot ( hatvalues ( finals.lm ), pch = 23 , bg = 'orange' , cex = 2 , ylab = 'Hat values' ) 
finals.table [ which ( hatvalues ( finals.lm )  >  0.3 ),]
```

```r
X = rnorm(100)
Y = 2 * X + 0.5 + rnorm(100)
alpha = 0.1
cutoff = qt(1 - alpha / 2, 97)
sum(abs(rstudent(lm(Y~X))) > cutoff)
```

```r
# Bonferroni correction
X = rnorm(100)
Y = 2 * X + 0.5 + rnorm(100)
cutoff = qt(1 - (alpha / 100) / 2, 97)
sum(abs(rstudent(lm(Y~X))) > cutoff)
```

```r
library(car)
outlierTest(finals.lm)
```
